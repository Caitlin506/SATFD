```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from hrv.detrend import smoothness_priors
```

Krótki wstęp 3-4 zdania

```python
#Wczytanie danych
chf201_1000=pd.read_csv("/content/sample_data/chf201.txt_N.csv",nrows=1000)
chf201_2000=pd.read_csv("/content/sample_data/chf201.txt_N.csv",nrows=2000)
chf201_5000=pd.read_csv("/content/sample_data/chf201.txt_N.csv",nrows=5000)
chf201_1000=np.array(chf201_1000)
chf201_2000=np.array(chf201_2000)
chf201_5000=np.array(chf201_5000)
```

```python
#plt.plot(chf201_1000)
```

```python
#Usuwanie niestacjonarnych trendów z wczytanych serii danych
rri1_detrended1 = smoothness_priors(rri=chf201_1000,l=500,fs=1)
rri1_detrended2 = smoothness_priors(rri=chf201_2000,l=500,fs=1)
rri1_detrended5 = smoothness_priors(rri=chf201_5000,l=500,fs=1)
```

```python
#Zaprezentowanie wyników na wykresach - porównanie danych przed i po obróbce
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,4))
ax1.set_title("Oryginalny")
ax2.set_title("Przetworzony")

ax2.plot(rri1_detrended1[0],color="blue")
ax2.plot(rri1_detrended2[0],color="red")
ax2.plot(rri1_detrended5[0],color="green")
ax1.plot(chf201_5000, color="black", label='original')
```

```python
chf202_1000=pd.read_csv("/content/sample_data/chf202.txt_N.csv",nrows=1000)
chf202_2000=pd.read_csv("/content/sample_data/chf202.txt_N.csv",nrows=2000)
chf202_5000=pd.read_csv("/content/sample_data/chf202.txt_N.csv",nrows=5000)
chf202_1000=np.array(chf202_1000)
chf202_2000=np.array(chf202_2000)
chf202_5000=np.array(chf202_5000)

rri1_detrended12 = smoothness_priors(rri=chf202_1000,l=500,fs=1)
rri1_detrended22 = smoothness_priors(rri=chf202_2000,l=500,fs=1)
rri1_detrended52 = smoothness_priors(rri=chf202_5000,l=500,fs=1)
```
