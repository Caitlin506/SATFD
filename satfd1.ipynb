```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from hrv.detrend import smoothness_priors
```

```python
chf201_1000=pd.read_csv("/content/sample_data/chf201.txt_N.csv",nrows=1000)
chf201_2000=pd.read_csv("/content/sample_data/chf201.txt_N.csv",nrows=2000)
chf201_5000=pd.read_csv("/content/sample_data/chf201.txt_N.csv",nrows=5000)
chf201_1000=np.array(chf201_1000)
chf201_2000=np.array(chf201_2000)
chf201_5000=np.array(chf201_5000)
```

```python
plt.plot(chf201_1000)
```

```python
rri1_detrended1 = smoothness_priors(rri=chf201_1000,l=500,fs=3)
rri1_detrended2 = smoothness_priors(rri=chf201_2000,l=500,fs=3)
rri1_detrended5 = smoothness_priors(rri=chf201_5000,l=500,fs=3)
plt.plot(rri1_detrended1[0],color="blue")
plt.plot(rri1_detrended2[0],color="red")
plt.plot(rri1_detrended5[0],color="green")
```

Problemy z usuwaniem trędów na końcach przedziału - pojawiają się wysokie piki.
