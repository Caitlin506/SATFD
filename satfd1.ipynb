```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from hrv.detrend import smoothness_priors
```

```python
chf201_1000=pd.read_csv("/content/sample_data/chf201.txt_N.csv",nrows=1000)
chf201_2000=pd.read_csv("/content/sample_data/chf201.txt_N.csv",nrows=2000)
chf201_5000=pd.read_csv("/content/sample_data/chf201.txt_N.csv",nrows=5000)
chf201_1000=np.array(chf201_1000)
chf201_2000=np.array(chf201_2000)
chf201_5000=np.array(chf201_5000)
```

```python
plt.plot(chf201_1000)
```

```python
rri1_detrended1 = smoothness_priors(rri=chf201_1000,l=500,fs=1)
rri1_detrended2 = smoothness_priors(rri=chf201_2000,l=500,fs=1)
rri1_detrended5 = smoothness_priors(rri=chf201_5000,l=500,fs=1)
```

```python
plt.plot(rri1_detrended1[0],color="blue",label='detrended - 1000 samples')
plt.plot(rri1_detrended2[0],color="red",label='detrended - 2000 samples')
plt.plot(rri1_detrended5[0],color="green",label='detrended - 5000 samples')
plt.plot(chf201_5000, color="black", label='original')
plt.legend()
```

Problemy z usuwaniem trdw na kocach przedziau - pojawiaj si wysokie piki.
